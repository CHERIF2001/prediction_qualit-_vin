### 3. Prétraitement

Il nous faut créer un dataframe d'entrainement. 

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('quality', axis=1)
y = df['quality']

seed = 33

# je dois créer une graine aléatoire pour m'assurer de la reproductibilité de mes données

X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state= seed, test_size= 0.2)

print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

Nous remarquons que nos données ne sont pas de la même échelle, il est important de les standardiser. 

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.fit_transform(X_test)

```

Avant de passer à la modélisation, on doit choisir une métrique de modélisation. Dans ce cas, on choisit la MSE ou l'erreur quadratique moyenne. 

```python
# pourquoi on fait une validation croisée ? Pourquoi on fait 
```

```python
# regression lineaire 
lr_cve = cross_val_score(
    LinearRegression(), 
    X_train_scaled,
    y_train,
    cv=3,
    scoring= "neg_root_mean_squared_error"
)


# random forest 

rf_cve = cross_val_score(
    RandomForestRegressor(random_state=seed), 
    X_train_scaled,
    y_train,
    cv=3,
    scoring= "neg_root_mean_squared_error"
)

# La régression à vecteurs de support

svm_cve = cross_val_score(
    SVR(), 
    X_train_scaled,
    y_train,
    cv=3,
    scoring= "neg_root_mean_squared_error"
)
```

```python
print(lr_cve, rf_cve, svm_cve)
```

Il est important de noter qu'on regarde l'*erreur*. Plus la valeur est grande plus le modèle n'est pas performant. 
l'erreur du random forest semble plus faible

```python
# je peux calculer la moyenne de chaque modèle:

print("LR: ", lr_cve.mean(), "RF: ", rf_cve.mean(), "SVR: ", svm_cve.mean())
```

Ce qui corrobore notre idée. Le modèle de forêt aléatoire semble plus performant. C'est ce modèle qu'on va utiliser dans le cadre de notre travail

## 4. Contruction du modèle

```python
model = RandomForestRegressor(random_state=seed)
model.fit(X_train_scaled, y_train)
```

```python
# on fait notre estimation
y_pred = model.predict(X_test_scaled)
y_pred
```

```python
mean_squared_error(y_test, y_pred, squared=False)
```

notre vraie valeur se situe entre [a-0.65, a+0.65] avec a= la qualité 

## 5. Conclusions 

```python
model.feature_importances_
```

```python
vars_imp = pd.Series(model.feature_importances_, index=X.columns).sort_values(ascending=False)
vars_imp
```

On peut conclure que que pour produire un vin rouge, les principaux paramètres à agir sont l'alcool, l'acidité volatile et le dioxide de soufre.
